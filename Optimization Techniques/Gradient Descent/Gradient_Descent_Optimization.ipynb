## Gradient Descent Algorithm 

How the Optimization is done using gradient descent method. It is one of the most widely used optimization method, from linear regression to deep learning it is widely used.

```python
# Importing Our Libraries

%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
plt.style.use("ggplot")
```

```python
# Generating Random Numbers
a = np.arange(20)
print(a)
```

```python
# Making our Function
def lf(a):
    return (a-10)**2
```

```python
loss = lf(a)
```

```python
# Plotting Our Function
plt.plot(loss)
plt.show()
```

```python
# Coding Our Gradient Descent Algorithm


error = []
def GD():
    a_ = 0
    lr = 0.1
    plt.plot(loss)
    
    
    
    
    for i in range(50):
        gradient = 2*(a_-10)
        
        y = (a_-10)**2
        
        error.append(y)
        
        # Using update rule to continue
        # Minimizing the gradient
        a_ = a_ - lr*gradient
        losss = lf(a_)
        plt.scatter(a_,losss)   # Plotting the loss and the optimal gradients
        
        
        # string formatting
        print(" A_: %.3f Loss:  %.3f "%(a_,losss))
        
GD()
```

Here we can see that how the error is decreasing with the different optimal values.
